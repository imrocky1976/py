```python
import pymysql

# 打开数据库连接
db = pymysql.connect("198.87.91.177","his","his1234","his" )
```

```python

# 使用 cursor() 方法创建一个游标对象 cursor
cursor = db.cursor()



sub = 10000
sql = "INSERT INTO analogvalue(attr_oid, name) VALUES ('{oid}', '{name}')"

try:
    # 执行sql语句
    cursor.execute('delete from analogvalue')
    db.commit()
    
    for i in range(2700): 
        # SQL 插入语句
        cursor.execute(sql.format(oid='66-' + str(sub) + '-1', name='test' + str(sub)))
        sub += 1
        print('insert oid: ' + '66-' + str(sub) + '-1')
    # 提交到数据库执行
    db.commit()
except Exception as e:
    # 如果发生错误则回滚
    db.rollback()
    print(e)
    
print('ok')
```
